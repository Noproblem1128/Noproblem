```python
import random

def number_lottery(num_choices, num_draws):
    """
    주어진 범위에서 랜덤 번호를 추첨합니다.

    Parameters:
    - num_choices: int, 선택할 수 있는 번호의 최대값.
    - num_draws: int, 추첨할 번호의 개수.

    Returns:
    - list of int: 추첨된 번호 리스트 (무작위 순서).
    """
    lottery_numbers = list(range(1, num_choices + 1))
    random.shuffle(lottery_numbers)  # 번호를 무작위로 섞음
    drawn_numbers = lottery_numbers[:num_draws]  # num_draws 개의 번호를 추출
    return drawn_numbers

def main():
    """
    사용자와 상호작용하여 랜덤 번호 추첨을 실행합니다.
    """
    while True:
        num_choices = int(input("선택할 수 있는 번호의 범위를 입력하세요: "))
        num_draws = int(input("몇 개의 번호를 뽑을까요?: "))

        if num_draws > num_choices:
            print("뽑을 수 있는 번호의 개수가 범위보다 큽니다.")
            continue  # 루프의 처음으로 되돌아감

        numbers = number_lottery(num_choices, num_draws)
        print("추첨된 번호는 다음과 같습니다:")
        print(numbers)

        play_again = input("다시 추첨하시겠습니까? (y/n): ")
        if play_again.lower() != 'y':
            break  # 루프 종료

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams['font.family'] = 'Malgun Gothic'

file_path = '경찰청_범죄 발생 시간대 및 요일_20191231 (1).csv'


data = pd.read_csv(file_path, encoding='euc-kr')


data.head()
```

```python

time_columns = ['0시00분-02시59분', '03시00분-05시59분', '06시00분-08시59분',
                '09시00분-11시59분', '12시00분-14시59분', '15시00분-17시59분',
                '18시00분-20시59분', '21시00분-23시59분']

time_data = data.melt(id_vars=['범죄대분류', '범죄중분류'], value_vars=time_columns,
                      var_name='시간대', value_name='건수')

plt.figure(figsize=(12, 6))
sns.lineplot(data=time_data, x='시간대', y='건수', hue='범죄중분류')
plt.title('시간대별 범죄 발생 건수')
plt.xticks(rotation=45)
plt.show()
```
